```python
img2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/img/1645277499965.jpg')
img1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/img/FB_IMG_1689677983736.jpg')
img3 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/img/received_2939930732982847.jpeg')
```

```python
# get dimensions of image
dimensions = img1.shape
# height, width, number of channels in image
height = img1.shape[0]
width = img1.shape[1]
channels = img1.shape[2]
print ('dimensions: ',dimensions)
print ('height : ',height)
print ('width : ',width)
print ('channels : ',channels)
fig = plt.figure(figsize=(10,5)) # กาหนดขนาดรูปที่จะ ํ plot
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
ax = fig.add_subplot(1,3,1) # กาหนด ํ row ที่จะ plot รูป
ax.imshow(img1) # ax รับรูป img ที่จะแสดง
ax.set_title('6306021620069') # ax กาหนด ํ title ให้กบรูป ั
#แทรกคําสัง โหลดรูปมาที่ ่ img2
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB) # กาหนดระบบสี ํ เป็ น RGB
ax2 = fig.add_subplot(1,3,2)
ax2.imshow(img2)
ax2.set_title("Chanoknan Khoonwat")

img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB) # กาหนดระบบสี ํ เป็ น RGB
ax2 = fig.add_subplot(1,3,3)
ax2.imshow(img3)
ax2.set_title("IT")
plt.show() # แสดงรูปใน plt
```
