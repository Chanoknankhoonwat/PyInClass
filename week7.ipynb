```python
#นําเข้า library
import numpy as np
import cv2
from matplotlib import pyplot as plt

#mount โฟลเดอร์ Google drive
from google.colab import drive
drive.mount('/content/drive/')
```

```python
#โหลดภาพ splash.jpg มาเกบไว้ใน ็ img
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/img/splash.jpg')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dimensions = img2.shape
print ('picture dimensions: ',dimensions)
```

```python
#สร้างหน้ากากตัวกรอง
mask = np.array([[-1,-1,1],[-1,8,-1],[-1,-1,-1]])
img_point = cv2.filter2D(img2,-1,mask)
T = 10
img_binary = img_point >= T
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img2,cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Point")
plt.imshow(img_point,cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Binary")
plt.imshow(img_binary,cmap='gray')

plt.show()
```
