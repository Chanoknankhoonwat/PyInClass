```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

from google.colab import drive
drive.mount('/content/drive/')
```

```python
haarcascade_frontalface_default = ('/content/drive/MyDrive/Colab Notebooks/xml/haarcascade_frontalface_default.xml')
# โหลดภาพที่มีใบหน้า
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/img/lena.png')
# เรียกใช้ cv2.CascadeClassifier
faceCascade = cv2.CascadeClassifier(haarcascade_frontalface_default)
img2 = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
# ค้นหาใบหน้าในภาพแล้ววาดกรอบ
faces = faceCascade.detectMultiScale(img)

for (x, y, w, h) in faces :
  cv2.rectangle(img2, (x, y), (x + w, y + h), (0, 255, 0), 5)
```

```python
# plot ภาพ
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,2,1)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
ax.imshow(img)
ax.set_title('Original')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img2)
ax2.set_title('Face Detection')
plt.show()
```

```python
# ภาพหลายหน้า
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/img/faces.jpg')

img2 = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
faceCascade = cv2.CascadeClassifier(haarcascade_frontalface_default)
faces = faceCascade.detectMultiScale(img)

for (x, y, w, h) in faces :

  cv2.rectangle(img2, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,2,1)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
ax.imshow(img)
ax.set_title('Original')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img2)
ax2.set_title('Face Detection')
plt.show()
print(f'There are {len(faces)} faces found.')
```
